# Exploratory Data Analysis and Live tracking of Covid Cases

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#reading data from the government website.
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
#deleting the last row as it is the total number and the 1st one as we have already index and we dont nned that
df=df.iloc[:-1,1:]
```

```python
df
```

```python
df.columns
```

```python
#positive cases
df["positive"]
```

```python
#maximum cases was given below
max(df["positive"])
```

```python
#arranging the cases as below as we use this inplotting the graph
total_states=np.arange(len(df["state_name"]))
```

```python
total_states
```

```python
#plotting the graph for positive according the state

plt.figure(figsize=(10,10))
plt.barh(total_states,df["positive"],color="blue",alpha=0.5)
plt.yticks(total_states,df["state_name"])
#plt.xlim(1,max(df["positive"])+100)
plt.ylabel("states")
plt.xlabel("positive cases")
plt.title("Positive cases in States in INDIA")
```

```python
#plotting the graph for active according the state

plt.figure(figsize=(10,10))
plt.barh(total_states,df["active"],color="blue",alpha=0.5)
plt.yticks(total_states,df["state_name"])
#plt.xlim(1,max(df["positive"])+100)
plt.ylabel("states")
plt.xlabel("active cases")
plt.title("active cases in States in INDIA")
```

```python
#plotting the graph for death cases according the state

plt.figure(figsize=(10,10))
plt.barh(total_states,df["death"],color="blue",alpha=0.5)
plt.yticks(total_states,df["state_name"])
#plt.xlim(1,max(df["positive"])+100)
plt.ylabel("states")
plt.xlabel("death cases")
plt.title("death cases in States in INDIA")
```

```python
#stacking the bar chart because to view every thing ina single graph and even we can compare with another feature for any feature

# making the state name as the index
df=df.set_index("state_name", drop=True)
```
