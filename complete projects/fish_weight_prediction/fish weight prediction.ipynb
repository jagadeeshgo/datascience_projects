```python
#we will take linear regression
model=LinearRegression()
```

```python
#we will split the data
x=data[["Species","Length1","Length3","Height","Width"]]
y=data["Weight"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=7)
```
