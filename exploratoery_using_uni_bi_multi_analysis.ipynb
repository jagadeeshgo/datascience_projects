```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("iris.data")
```

```python
df.head()
```

```python
#uniivariate analysis
setosa=df.loc[df["Species"]=="Iris-setosa"]
```

```python
versicolor=df.loc[df["Species"]=="Iris-versicolor"]
```

```python
virginica=df.loc[df["Species"]=="Iris-virginica"]
```

```python
plt.plot(setosa["SepalLength"],np.zeros_like(setosa["SepalLength"]),"o")
plt.plot(versicolor["SepalLength"],np.zeros_like(versicolor["SepalLength"]),"o")
plt.plot(virginica["SepalLength"],np.zeros_like(virginica["SepalLength"]),"o")
```

```python
#Bivariate analysis
sns.FacetGrid(df,hue="Species",size=7).map(plt.scatter,"SepalLength","SepalWidth").add_legend()
```

```python
#multi variate analysis
sns.pairplot(df,hue="Species")
```

```python
train,test=train_test_split(df,test_size=0.2,random_state=7)
```
